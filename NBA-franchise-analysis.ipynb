```python
import csv
import pandas as pd
import string
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
file_path = 'NBA-Complete.csv'

NBA_df = pd.read_csv(file_path)
NBA_df.fillna(0)
```

```python
X = NBA_df['Team']
Y2017= NBA_df['Attend. (2017)']
Y2019= NBA_df['Attend. (2019)']
#Y2019= NBA_df['Pct (2019)']
#Y2020= NBA_df['Pct (2020)']
#Y2021= NBA_df['Pct (2021)']

figure(figsize=(300, 100), dpi=20)
plt.rc('font', size=10) #controls default text size

  
X_axis = np.arange(len(X))

fig, ax1 = plt.subplots()
  
plt.bar(X_axis - 0.2, Y2017, 0.3)


plt.bar(X_axis + 0.2, Y2019, 0.3)
#plt.bar(X_axis, Y2019, 0.2)
#plt.bar(X_axis + 0.2, Y2020, 0.2)
#plt.bar(X_axis + 0.4, Y2021, 0.2)
  
plt.xticks(X_axis, X, rotation=90)
plt.xlabel("Team")
plt.ylabel("Attendance")
```

```python
X = NBA_df['Team']
Y2017_W= NBA_df['Pct (2017)']
Y2019_W= NBA_df['Pct (2019)']

figure(figsize=(300, 100), dpi=20)
plt.rc('font', size=10)

X_axis = np.arange(len(X))

fig, ax1 = plt.subplots()
  
plt.bar(X_axis - 0.2, Y2017_W, 0.3)
plt.bar(X_axis + 0.2, Y2019_W, 0.3)

plt.xticks(X_axis, X, rotation=90)
plt.xlabel("Team")
plt.ylabel("Win Pct")
```
